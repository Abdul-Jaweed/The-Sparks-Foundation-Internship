### Sub-Category wise Sales, Profit and Discount Analysis

```python
sub_cat = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']]
```

```python
sub_cat = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
```

```python
sub_cat.head()
```

```python
plt.figure(figsize = (25,25))
plt.pie(sub_cat['Sales'], labels = sub_cat.index, autopct = '%1.1f%%')
plt.title('Sub-Category wise Sales, Profit and Discount Analysis', fontsize = 35)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (25,25))
plt.pie(sub_cat['Discount'], labels = sub_cat.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 35)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

```python
sub_cat.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                       figsize= (20,10),
                                                       label=['Average Sales Price($)','Profit($)'])
plt.title('Sub-Category Wise Profit Analysis', fontsize = 35)
plt.legend()
plt.show()
```
